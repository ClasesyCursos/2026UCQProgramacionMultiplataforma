## ¿Qué es React Native?
React Native permite desarrollar aplicaciones para Android, iOS y Web utilizando JavaScript/TypeScript y React.

## El problema antes de Expo
La configuración tradicional requería Android Studio, SDKs, Java, emuladores y configuración manual.

## ¿Qué es Expo?
Expo es un framework construido sobre React Native que simplifica el desarrollo mediante herramientas, SDK y configuración automática.

## Arquitectura
```text
Aplicación
   ↓
React Native
   ↓
Expo SDK
   ↓
Android / iOS
```

## Expo SDK
Incluye APIs para cámara, GPS, sensores, archivos, notificaciones, Bluetooth, audio y más.

## Expo Go
Permite ejecutar aplicaciones escaneando un código QR sin generar un APK.

## Flujo de desarrollo
```text
VS Code → npm start → Expo Server → QR → Expo Go → Aplicación
```

## Expo vs React Native CLI
|Expo|CLI|
|---|---|
|Configuración sencilla|Manual|
|Expo Go|No|
|OTA|Sí|
|Ideal para aprender|Sí|
